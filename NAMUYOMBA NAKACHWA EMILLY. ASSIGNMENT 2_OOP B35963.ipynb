#QUESTION ONE : ELECTRICITY BILLING OPTIMIZATION

```python

# QUESTION ONE: ELECTRICITY BILLING OPTIMIZATION

print("*" * 100)
print("QUESTION ONE:ELECTRICITY BILL OPTIMIZATION")
print("*" * 100)
print()

# Analyzing monthly household electriicity consumption data for 10 customers in Kampala.
# List of Customer names
customer_names=['Bosco','Jane','Sonia','Alex','Sarah','Brain','Emilly','Loyce','Eric','Paul']
# List of units used by each customer.
units_used=[245,450,510,80,600,300,150,200,480,100]
# List of connection types ,whoich detrmines the billing rate
connection_type=['Domestic','Domestic','Commercial','Domestic','Commercial','Domestic','Commercial','Domestic','Commercial','Domestic']
# The billing rates as specified in the assignment
DOMESTIC_RATE=820 # UGX/unit
COMMERCIAL_RATE=1000 #UGX/unit

customer_data=list(zip(customer_names,units_used,connection_type))
print("Sample customer Data (Name,Units,Type):")
print(customer_data)
print("-"*50)
#-----------------------------------------------------------------------------
#1.1. Calculate Bills using a lamba and map()
#-----------------------------------------------------------------------------

calculate_bill=lambda units,connection_type:units * DOMESTIC_RATE if connection_type=='Domestic' else units * COMMERCIAL_RATE
# Use map() to apply lamba function to all customer data.
# We extarct the units (item[1]) and connection type (item[2] from each tuple) in customer data.
total_bills=list(map(
   lambda customer:calculate_bill(customer[1],customer[2]),
   customer_data
 ))
print("List of calculated Bills (UGX):")
print(total_bills)

name_bill_pairs=list(zip(customer_names,total_bills))
print("Customer Name and Total Bill Pairs:")
print(name_bill_pairs)


#-------------------------------------------------------------------------
# 1.2 Use filter() for High-Consumption Customers
#-------------------------------------------------------------------------

# Use a simple lamba to check if consuption is > 500 units.
is_high_consumer=lambda data:data[1]> 500


high_consumption_customers=list(filter(is_high_consumer,customer_data))
print("Customers with consumption >500 units:")
# The result is a list of tuples containing (Name,Units,Type)


#-------------------------------------------------------------------------
# 1.3 .Function using*args to compute average
#-------------------------------------------------------------------------

def average_bill(*bills):
    """Computes the average of all bills passed as arguments."""
    
    if not bills:
        return 0
    

    return sum(bills)/len(bills)


print(f"The total number of bills included in the average is: {len(total_bills)}")
print("\nAverage bill: UGX", round(average_bill(*total_bills), 2))


#-----------------------------------------------------------------------------------
# 1.4 Display results using list comprehension.
#-----------------------------------------------------------------------------------
# List comprehension syntax:[expression for item in iterable]
# We iterate over the 'name_bills_pais'(tuples) and create a formateed string for each pair
final_summary=[
    f"{name}:UGX{bill:,}"
    for name,bill in name_bill_pairs
]
print("Final Formatted Billing Summary:")
print(final_summary)

```

```python

# MARKET BASKET PRICE AGGREGATOR
from itertools import zip_longest
fruit_names=['mango','orange','apple']
mango_prices=[2500, 2700, 2600, 2800]
orange_prices=[3000,3200,3100,3050]
apple_prices=[4500, 4600, 4550]

print("-" * 50)
print(" 2.1 Use longest () to handle missing prices")
print("-" * 50)
#-------------------------------------------------
# 2.1 Use z_longest() to handle missing prices.
#-------------------------------------------------

# zip_longest fills missing values (Vendor 4 for apple_prices) with a specified fillvalue.
# Using 0 is simple for a beginner, but note that it skews the average.

all_prices_padded=list(zip_longest(mango_prices, orange_prices, apple_prices, fillvalue=0))
prices_by_fruit=list(zip(*all_prices_padded))
print("2.1 Prices by Vendor (Padded with 0 for missing data):")
print(all_prices_padded)

print("-" * 50)
print(" 2.2 Define lambda for average")
print("-" * 50)

# The lambda computes sum/length for any list passed to it
avg=lambda lst: sum(lst)/len(lst)

average_prices=list(map(avg,prices_by_fruit))
print(f"Average Prices:{list(zip(fruit_names, average_prices))}")

print("-" * 50)

#----------------------------------------------------------------------------------
# 2.3: Create a generator expression
#----------------------------------------------------------------------------------


fruits_above_avg=(
    name for name, price in zip(fruit_names,average_prices)
    if price > 3000
)
print("2.3 Generator creayed butn not consumed.")



# ------------------------------------------------------------------------------------
# 2.4. Print the generator 's resuts neatly
#-------------------------------------------------------------------------------------
results_name=list(fruits_above_avg)

output_string="Fruits Above Average Price: "+" , ". join(results_name)

print(output_string)
```


#QUESTION THREE:DISTRICT TEMPERATURE TRACKER


```python

MONTHS = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
kampala_celsius = [28, 30, 29, 27, 26, 29, 30, 31, 32, 30, 29, 28]

# ----------------------------------------------------
# 3.1. Use map() with lambda to convert to Fahrenheit
# ----------------------------------------------------

c_to_f = lambda c: (c * 9/5) + 32

kampala_fahrenheit = list(map(c_to_f, kampala_celsius))

print("3.1 Kampala Temperatures in Fahrenheit:")

print(list(zip(kampala_celsius, kampala_fahrenheit)))


# ----------------------------------------------------
# 3.2. Use zip() to pair months with Kampala temps
# ----------------------------------------------------

kampala_monthly_temps = list(zip(MONTHS, kampala_celsius))

print("3.2 Kampala Monthly Temps (Month, °C):")
print(kampala_monthly_temps)

# ----------------------------------------------------
# 3.3. Use filter() to identify hot months (> 30°C)
# ----------------------------------------------------

is_hot_month = lambda month_temp: month_temp[1] > 30

hot_months_list = list(filter(is_hot_month, kampala_monthly_temps))

print("3.3 Hot Months List (> 30°C):")
print(hot_months_list)

# ----------------------------------------------------
# 3.4. Create a generator expression
# ----------------------------------------------------

# Generator yields month names and temperatures that are above 30°C
hot_months_generator = (
    (month, temp) for month, temp in kampala_monthly_temps
    if temp > 30
)

print("3.4 Generator created.")

# ----------------------------------------------------
# 3.5. Print generator results neatly
# ----------------------------------------------------

formatted_hot_months = [
    f"{month} - {temp}°C"
    for month, temp in hot_months_generator
]
output_string = "Hot months in Kampala: " + ", ".join(formatted_hot_months)

print(output_string)
```

QUESTION 4:SCHOOL FEES PAYMENT ANALYZER

```python




students = ['Alex', 'Grace', 'Sarah', 'Brian']

installments = [
    [150000, 200000, 250000],
    [500000, 0, 200000], 
    [300000, 300000, 300000],
    [400000, 100000, 0] 
]

# ----------------------------------------------------
# 4.1. Lambda function to sum valid payments (ignores zeros)
# ----------------------------------------------------


sum_valid_payments = lambda payments: sum(p for p in payments if p > 0)

# ----------------------------------------------------
# 4.2. Use map() to compute total paid by each student
# ----------------------------------------------------

total_paid_by_student = list(map(sum_valid_payments, installments))

print("4.2 Total Paid by Students:")
print(total_paid_by_student)

# ----------------------------------------------------
# 4.3. Combine student names and totals using zip()
# ----------------------------------------------------

student_payment_pairs = list(zip(students, total_paid_by_student))

print("4.3 Student (Name, Total Paid):")
print(student_payment_pairs)



# ----------------------------------------------------
# 4.4. Use filter() to find students who cleared full fees (>= 600,000)
# ----------------------------------------------------

# Lambda checks if the total paid (item[1] in the tuple) is >= 600000
cleared_full_fees = lambda student_pair: student_pair[1] >= 600000

# Use filter() on the combined pairs
full_fee_cleared_students = list(filter(cleared_full_fees, student_payment_pairs))

print("4.4 Students who cleared full fees (>= 600,000):")
print(full_fee_cleared_students)
# ----------------------------------------------------
# 4.5. Create payment_summary(**kwargs)
# ----------------------------------------------------

# **kwargs captures named arguments into a dictionary
def payment_summary(**payments):
    """Receives named arguments (student=total) and prints summaries."""
    print("4.5 Payment Summary:")
    for student, total in payments.items():
        print(f"Summary for {student}: Total paid is UGX {total:,}.")


payment_dict = dict(student_payment_pairs)


payment_summary(**payment_dict)

```

QUESTION 5:AGRICULTURAL YIELD ESTIMATOR

```python

districts = ['Bushenyi', 'Mityana', 'Kasese', 'Mbale']
yield_data_kg = [1200, 1500, 900, 1300] # Yield in kg [cite: 56]

# ----------------------------------------------------
# 5.1. List comprehension with a lambda (KG to Tons)
# ----------------------------------------------------

kg_to_ton = lambda kg: kg / 1000

# List comprehension applies the lambda to each yield data point
yield_data_tons = [kg_to_ton(kg) for kg in yield_data_kg]

# Combine districts and yields in tons
district_yield_pairs = list(zip(districts, yield_data_tons))

print("5.1 Yield Data (District, Tons):")
print(district_yield_pairs)


# ----------------------------------------------------
# 5.2. Generator that yields districts with yield > 1 ton
# ----------------------------------------------------

# Generator expression yields the district name if the yield (item[1]) is > 1.0
high_yield_districts_generator = (
    district for district, tons in district_yield_pairs
    if tons > 1.0
)

print("5.2 Districts with Yield > 1 Ton (from Generator):")
print(list(high_yield_districts_generator))


# ----------------------------------------------------
# 5.3. Use *args to compute average yield across all districts
# ----------------------------------------------------

def average_yield(*yields):
    """Computes the average yield from a variable number of yield values."""
    if not yields:
        return 0
    return sum(yields) / len(yields)

avg_yield_tons = average_yield(*yield_data_tons)

print(f"5.3 Average Yield Across All Districts: {avg_yield_tons} tons")

# ----------------------------------------------------
# 5.4. Use **kwargs to simulate price and compute revenue (tons * price)
# ----------------------------------------------------

def compute_revenue(yields_in_tons, **prices):
    """Computes revenue (tons * price) based on district prices."""
    print("5.4 Revenue Analysis:")
    revenue_summary = {}
    for district, yield_t in yields_in_tons:
        # Get the price for the district from the **prices dictionary, defaulting to 0
        price = prices.get(district, 0)
        revenue = yield_t * price
        revenue_summary[district] = revenue
        print(f"  {district} Revenue: UGX {revenue:,.0f} (at UGX {price:,}/ton)")
    return revenue_summary

district_prices = {
    'Bushenyi': 500000,
    'Mityana': 480000,
    'Kasese': 550000,
    'Mbale': 490000
}

revenue_results = compute_revenue(district_yield_pairs, **district_prices)
print("Revenue Calculation Results. generating final formatted output....")

    
# ----------------------------------------------------
# 5.5. Print formatted output (e.g., Kasese produced 0.9 tons)
# ----------------------------------------------------

print("\n5.5 Formatted Yield Output:")
for district, tons in district_yield_pairs:
    revenue= revenue_results.get(district,0)
out_put = (
        f"{district} produced {tons} tons — "
        f"Revenue: UGX {revenue:,.0f}"
    )
print(out_put)

```

#QUESTION 6:WEB DATA AGGREGATION

```python


import requests
sites = ['https://ucu.ac.ug', 'https://harba.ug', 'https://www.bou.or.ug', 'https://badurl.org'] # Added a bad URL for demo

# ----------------------------------------------------
# 6.1. Function with *urls to return response codes (SIMULATED)
# ----------------------------------------------------


def get_response_codes(*urls):
    """Simulates taking multiple URLs and returning their response codes."""
    # SIMULATED RESPONSE CODES: Assuming ucu.ac.ug and bou.or.ug are 200 (Active)
    # and harba.ug and badurl.org are 404 (Not Found) or similar.
    codes = []
    for url in urls:
        if 'ucu.ac.ug' in url or 'bou.or.ug' in url:
            codes.append(200) # Active
        else:
            codes.append(404) # Not Found
    return codes

# Unpack the list of sites using *
status_codes = get_response_codes(*sites)
url_code_pairs = list(zip(sites, status_codes))

print("6.1 URL Status Codes (Simulated):")
print(url_code_pairs)

# ----------------------------------------------------
# 6.2. List comprehension to print URLs with status code 200
# ----------------------------------------------------

active_urls_list = [
    url for url, code in url_code_pairs
    if code == 200
]

print("6.2 Active URLs (Status Code 200) via List Comprehension:")
print(active_urls_list)

# ----------------------------------------------------
# 6.3. Store results in a dictionary using a dictionary comprehension
# ----------------------------------------------------


status_dict = {
    url: code for url, code in url_code_pairs
}

print("6.3 URL Status Dictionary via Dictionary Comprehension:")
print(status_dict)
print("-" * 50)

# ----------------------------------------------------
# 6.4. Create a generator expression for active sites
# ----------------------------------------------------

active_site_generator = (
    f"Active Site: {url}"
    for url, code in url_code_pairs
    if code == 200
)

print("6.4 Active Site Generator Results:")
for message in active_site_generator:
    print(message)
```
